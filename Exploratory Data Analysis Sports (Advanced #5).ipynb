```python
df2=df2.replace(np.nan, 0)
df2.isnull().sum()
```

# Most Successful team

```python
print(df2['winner'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df2['winner'])
plt.xticks(rotation=90)
plt.show()
```

# Most Successful Player

```python
print(df2['player_of_match'].value_counts())
plt.figure(figsize=(35,18))
sns.countplot(x=df2['player_of_match'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(df2['toss_winner'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df2['toss_winner'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(18,100))
sns.countplot(y='batsman',hue='total_runs', data=df1, palette="Greens_d")
plt.xticks(rotation=90)
plt.show()
```

```python
print(df2['toss_decision'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df2['toss_decision'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(35,18))
sns.countplot(x='toss_winner',hue='toss_decision', data=df2, palette="Greens_d")
plt.xticks(rotation=90)
plt.show()
```
